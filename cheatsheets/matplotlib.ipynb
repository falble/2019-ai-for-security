# Matplotlib

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. You can generate plots, histograms, power spectra, bar charts, errorcharts, scatterplots, etc., with just a few lines of code.

Useful links:
- [website](https://matplotlib.org/)
- [matplotlib.pyplot documentation](https://matplotlib.org/api/pyplot_summary.html)
- [sample plots](https://matplotlib.org/tutorials/introductory/sample_plots.html)

IMPORTANT: this is **not** a comprehensive list.

---

```python
import matplotlib.pyplot as plt

# numpy is used here only to create artificial data to visualize
import numpy as np
from numpy.random import normal,rand
```

## Line plot

```python
# this two lines create (artificial) data to visualize
a = np.linspace(0, 10, 100)
b = np.exp(-a)

# this is the actual visualization
plt.plot(a, b)
plt.show()
```

## Histogram

Some of the parameters you can play with:
- bins = the number of bins to consider while plotting the histogram
- c = the colour of the plot

```python
# create data
x = normal(size=200)

# visualization
plt.hist(x, bins=30)
plt.show()
```

## Scatter plot

Some of the parameters you can play with:
- c = the color of the scatter plot, you can either select a unique color for the whole scatter or one colour for each point
- s = the color of the scatter plot, you can either select a unique size for the whole scatter or one size for each point
- alpha = how transparent the scatter plot should be

```python
# create data
a = rand(100)
b = rand(100)

# perform the visualization
plt.scatter(a, b)
plt.show()
```

## Bar plot

```python
# create data
a = ['A', 'B', 'C']
b = [100, 10, 200]

# perform the visualization
plt.bar(a, b)
plt.show()
```

# Some more complex examples

## Plotting several lines on the same plot

```python
# this two lines create (artificial) data to visualize
x = np.linspace(0, 10, 100)
y1 = np.exp(-x)
y2 = x*1/20

# below, the actual visualization

# this row lets you initialize the visualization. Between here and the final plt.show() you can defined the 
#    characteristics of the visualization you want
fig, ax = plt.subplots()

ax.plot(x, y1, label='line 1')
ax.plot(x, y2, label='line 2')

ax.legend()  # to display the legend containing the two labels defined above
ax.set_title('a sample plot')  # display a name for the plot (it is on the top)

ax.set_xlabel('x axis') # give a name to the x axis
ax.set_ylabel('y axis') # give a name to the y axis

# to plot the visualization
plt.show()
```

As showed above, you can plot several items in the same plot, for instance two lines, a line and a scatter plot, etc.
